```python
import pandas as pd
```

```python
df = pd.read_csv('pharmacologically_active.csv')
df = df[df['PDB ID'].notna()]
len(df)
```

```python
df
```

```python
lst_protein = [protein.split(';')[0] for protein in df['PDB ID']] 
```

```python
dti_dataset = {'PDB ID': lst_protein, 'Drug IDs': df['Drug IDs']}
dti_dataset = pd.DataFrame(dti_dataset)

# Display the DataFrame
dti_dataset.to_csv('dti_dataset.csv')
```
