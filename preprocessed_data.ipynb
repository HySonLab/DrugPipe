```python
def add_chain_to_pdb(source_pdb, destination_pdb, new_chain_id):
    with open(source_pdb, 'r') as file:
        lines = file.readlines()
    
    with open(destination_pdb, 'w') as file:
        for line in lines:
            if line.startswith('ATOM') or line.startswith('HETATM'):
                # Change the chain identifier to the new one
                line = line[:21] + new_chain_id + line[22:]
            file.write(line)
    
    print(f'Chain {new_chain_id} added to {destination_pdb}')


folder_old = 'dude'
folder_new = 'dude_process'
for path in glob(f'/cm/archive/phucpht/Drug_research/Blind_pocket/datasets/{folder_old}/*'):
    protein_name = os.path.split(path)[1]
    path = f'/cm/archive/phucpht/Drug_research/Blind_pocket/datasets/{folder_old}/{protein_name}'
    path_new = f'/cm/archive/phucpht/Drug_research/Blind_pocket/datasets/{folder_new}/{protein_name}'
    if not os.path.exists(path_new):
        os.makedirs(path_new)
    source_pdb_file = glob(f'{path}/*.pdb')[0]
    destination_pdb_file = os.path.join(path_new, f'{protein_name}.pdb')
    new_chain_identifier = 'A'  # New chain identifier

    add_chain_to_pdb(source_pdb_file, destination_pdb_file, new_chain_identifier)
    smile_list = []
    active     = [] 
    with open(f'{path}/actives_final.ism','r') as file:
        for line in file:
            smile_list.append(line.split(' ')[0])
            active.append(1)


    smile_list = []
    with open(f'{path}/decoys_final.ism','r') as file:
        for line in file:
            smile_list.append(line.split(' ')[0])
            active.append(0)

    
    combined = list(zip(smile_list, active))
    random.shuffle(combined)
    shuffled_list1, shuffled_list2 = zip(*combined)
    shuffled_list1 = list(shuffled_list1)
    shuffled_list2 =list(shuffled_list2)
    indices = list(range(len(shuffled_list1)))
    data = {'smiles': shuffled_list1, 'active': shuffled_list2, 'key_idx':indices}
    df = pd.DataFrame(data)
    df.to_csv(os.path.join(path_new,f'{protein_name}.csv'), index = False)
    
```

```python

# Usage
source_pdb_file = 'datasets/dude/adrb2/receptor.pdb'
destination_pdb_file = 'destination.pdb'
new_chain_identifier = 'A'  # New chain identifier

add_chain_to_pdb(source_pdb_file, destination_pdb_file, new_chain_identifier)
```
